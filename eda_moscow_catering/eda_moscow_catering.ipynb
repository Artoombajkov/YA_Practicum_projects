## Данные

Датасет с заведениями общественного питания Москвы, составленный на основе данных сервисов Яндекс Карты и Яндекс Бизнес на лето 2022 года.

- `rest_info.csv` - информация о заведениях общественного питания
- `rest_price.csv` - информация о среднем чеке в заведениях общественного питания


### Описание датасета `rest_info.csv`

- `id` - уникальный индификатор;
- `name` — название заведения;
- `address` — адрес заведения;
- `district` — административный район, в котором находится заведение, например Центральный административный округ;
- `category` — категория заведения, например «кафе», «пиццерия» или «кофейня»;
- `hours` — информация о днях и часах работы;
- `rating` — рейтинг заведения по оценкам пользователей в Яндекс Картах (высшая оценка — 5.0);
- `chain` — число, выраженное 0 или 1, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки):
  
        0 — заведение не является сетевым;
        1 — заведение является сетевым.
- `seats` — количество посадочных мест.


### Описание датасета `rest_price.csv`

- `id` - уникальный индификатор;
- `price` — категория цен в заведении, например «средние», «ниже среднего», «выше среднего» и так далее;
- `avg_bill` — строка, которая хранит среднюю стоимость заказа в виде диапазона, например:
  
        «Средний счёт: 1000–1500 ₽»;
        «Цена чашки капучино: 130–220 ₽»;
        «Цена бокала пива: 400–600 ₽».
        и так далее;


- `middle_avg_bill` — число с оценкой среднего чека, которое указано только для значений из столбца `avg_bill`, начинающихся с подстроки «Средний счёт»:
Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт.
- `middle_coffee_cup` — число с оценкой одной чашки капучино, которое указано только для значений из столбца `avg_bill`, начинающихся с подстроки «Цена одной чашки капучино»:
Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», то в столбец ничего не войдёт.

Создадим столбец `is_24_7` с обозначением того, что заведение работает ежедневно и круглосуточно, то есть 24/7:
- логическое значение True — если заведение работает ежедневно и круглосуточно;
- логическое значение False — в противоположном случае.

```python
rest_all_info['is_24_7']= False
```

```python
def check_hours_and_set_is_24_7(df, hours_col):
    """
    Проверяет значение ячейки в столбце 'hours' на равенство "ежедневно, круглосуточно"
    и возвращает True в столбец 'is_24_7', если условие выполняется, и False в противном случае.

    Args:
        df (pd.DataFrame): DataFrame с данными.
        hours_col (str): Название столбца с информацией о часах работы.

    """
    for index, row in df.iterrows():
        if pd.isna(row[hours_col]):
           continue
        if row[hours_col] == "ежедневно, круглосуточно":
          df.loc[index, 'is_24_7'] = True
        
    return df
```

```python
hours_column = 'hours'
check_hours_and_set_is_24_7(rest_all_info, hours_column)
```

- Был добавлен столбец с локическими значениями о работе заведения (круглосуточно - True, или нет - False)

```python
# 3. Соотношение сетевых и несетевых заведений
# Считаем общее количество заведений
total_establishments = len(rest_all_info)

# Считаем количество сетевых и несетевых заведений
network_status_counts = rest_all_info['chain'].value_counts()

# Считаем долю сетевых и несетевых заведений
network_status_percentages = (network_status_counts / total_establishments) * 100

print(network_status_percentages)

print("\nСоотношение сетевых и несетевых в разрезе категорий (доли):")

# Группируем по категориям и считаем количество сетевых и несетевых в каждой
network_status_by_category = rest_all_info.groupby('category')['chain'].value_counts().unstack(fill_value=0)

# Считаем общее количество заведений в каждой категории
category_totals = network_status_by_category.sum(axis=1)

# Считаем доли сетевых и несетевых заведений в каждой категории
network_status_percentages_by_category = network_status_by_category.div(category_totals, axis=0) * 100

print(network_status_percentages_by_category.round(1))
# Визуализация долей
network_status_percentages_by_category.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Доли сетевых и несетевых заведений в разрезе категорий')
plt.xlabel('Категория')
plt.ylabel('Доля заведений (%)')
plt.legend(title='Сетевой статус', labels=['Несетевые', 'Сетевые'])
plt.xticks(rotation=45, ha='right')  # Поворот подписей оси x
plt.tight_layout()
plt.show()
```

Сетевых заведений больше всего у "кафе". Пицеррий и кофеен почти поровну между сетевыми и не сетевыми (разница в 27 штук в пользу сетевых)

Приведем все столбцы датасетов к стилю snake_case

```python
def convert_to_snake_case(name):
    """Преобразует строку в snake_case."""
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower().replace(' ', '_').replace('__','_')

# Итерируемся по словарю и преобразуем названия столбцов
for table_name, df in all_tables.items(): 
    # Преобразуем названия столбцов
    new_columns = {col: convert_to_snake_case(col) for col in df.columns}
    df.rename(columns=new_columns, inplace=True)
    
    print("Новые названия столбцов:")
    print(df.columns.tolist())
```

### Проверим каждую таблицу на наличе пропусков

```python
# Функция для проверки пропущенных значений
def check_missing_values(table_name, df, column_descriptions):
    """
    Проверяет каждый столбец DataFrame на наличие пропущенных значений,
    выводит количество пропусков, их процент и описание столбца.

    Args:
        table_name (str): Название таблицы
        df (pd.DataFrame): DataFrame для анализа.
        column_descriptions (dict): Словарь описаний столбцов.
    """
    print(f"\nАнализ пропущенных значений для таблицы: '{table_name}'")
    for column in df.columns:
        missing_count = df[column].isnull().sum()
        total_count = len(df)
       
        description = column_descriptions.get(column, 'Описание отсутствует')
        missing_percentage = (missing_count / total_count) * 100

        # Убедимся, что missing_percentage является скаляром
        if isinstance(missing_percentage, (int, float)):
            if missing_percentage > 0:
                print(f"  - Столбец: {column}. Описание: {description}.  Пропуски: {missing_count}. Процент пропусков: {missing_percentage:.2f}% \n")
                
        else:
            # Если missing_percentage является Series, обработаем его поэлементно
            for idx, value in missing_percentage.items():
                if value > 0:
                    print(f"  - Столбец: {column} (строка {idx}). Описание: {description}. Пропуски: {missing_count}. Процент пропусков: {value:.2f}%\n")

# Применяем функцию ко всем таблицам
for table_name, table_info in data_dictionary.items():
    if table_info['df'] is not None:  # Проверяем, что датафрейм существует
        check_missing_values(table_name, table_info['df'], table_info['columns'])
    else:
        print(f"\nТаблица {table_name} не загружена (df = None).")
```

## Шаг 2. Предобработка данных, предварительное исследование

### 2.1. Раунды финансирования по годам

```python
# Фильтрация данных по годам с более чем 50 раундами
company_and_rounds['year'] = company_and_rounds['funded_at'].dt.year
rounds_by_year = company_and_rounds.groupby('year').agg(
    typical_raised_amount=('raised_amount', 'median'),
    total_rounds=('raised_amount', 'count')
).reset_index()

# Фильтрация по количеству раундов
rounds_by_year = rounds_by_year[rounds_by_year['total_rounds'] > 50]

# Визуализация
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(rounds_by_year['year'], rounds_by_year['typical_raised_amount'], marker='o')
plt.title('Динамика типичного размера средств в раундах финансирования')
plt.xlabel('Год')
plt.ylabel('Типичный размер средств')
plt.grid(True)
plt.show()

# Ответы на вопросы
max_year = rounds_by_year.loc[rounds_by_year['typical_raised_amount'].idxmax(), 'year']
print(f"Год с максимальным типичным размером средств: {max_year}")

# Анализ 2013 года
year_2013 = rounds_by_year[rounds_by_year['year'] == 2013]
print(f"Данные за 2013 год:\n{year_2013}")
```

### 2.2. Люди и их образование

```python
# Объединение данных о людях и их образовании
people_education = pd.merge(people, education, left_on='id', right_on='person_id', how='left')

# Группировка по компаниям и расчет доли пропусков в образовании
people_education['has_education'] = people_education['instituition'].notna()
education_by_company = people_education.groupby('company_id').agg(
    total_employees=('id', 'count'),
    employees_without_education=('has_education', lambda x: (x == False).sum())
).reset_index()

education_by_company['missing_education_ratio'] = education_by_company['employees_without_education'] / education_by_company['total_employees']

# Разделение компаний на группы по количеству сотрудников
bins = [0, 10, 50, 100, float('inf')]
labels = ['0-10', '11-50', '51-100', '100+']
education_by_company['employee_group'] = pd.cut(education_by_company['total_employees'], bins=bins, labels=labels)

# Средняя доля сотрудников без информации об образовании
mean_missing_education = education_by_company.groupby('employee_group')['missing_education_ratio'].mean()
print(mean_missing_education)
```
